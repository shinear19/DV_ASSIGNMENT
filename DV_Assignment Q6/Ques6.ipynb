```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Set the aesthetics of the plots
sns.set(style='whitegrid', palette='pastel', font_scale=1.2)

# Create a figure
plt.figure(figsize=(12, 6))

# 1. Scatter Plot with Aesthetics
plt.subplot(1, 2, 1)
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species', style='species', s=100)
plt.title('Scatter Plot of Sepal Dimensions')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')

# 2. Box Plot with Aesthetics
plt.subplot(1, 2, 2)
sns.boxplot(data=iris, x='species', y='petal_length', palette='Set2')
plt.title('Box Plot of Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')

# Show the plots
plt.tight_layout()
plt.show()
```
