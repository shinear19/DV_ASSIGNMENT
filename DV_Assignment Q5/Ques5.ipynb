```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)  # 100 points between 0 and 10
y = np.sin(x)  # Sine wave
y2 = np.cos(x)  # Cosine wave
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 1, 8]
data = [15, 30, 45, 10]  # For pie chart (ensure these are non-negative)

# 1. Line Plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Sine Wave', color='blue', linestyle='-', linewidth=2)
plt.plot(x, y2, label='Cosine Wave', color='orange', linestyle='--', linewidth=2)
plt.title('Line Plot of Sine and Cosine Waves')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid()
plt.show()

# 2. Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='green', marker='o', label='Sine Points')
plt.scatter(x, y2, color='red', marker='x', label='Cosine Points')
plt.title('Scatter Plot of Sine and Cosine Waves')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid()
plt.show()

# 3. Bar Chart
plt.figure(figsize=(10, 6))
plt.bar(categories, values, color='purple')
plt.title('Bar Chart Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.grid(axis='y')
plt.show()

# 4. Histogram
plt.figure(figsize=(10, 6))
data_hist = np.random.randn(1000)  # Random data for histogram
plt.hist(data_hist, bins=30, color='cyan', edgecolor='black')
plt.title('Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()

# 5. Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(data, labels=['Category 1', 'Category 2', 'Category 3', 'Category 4'], autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart Example')
plt.axis('equal')  # Equal aspect ratio ensures pie is drawn as a circle.
plt.show()
```
