```python
import pandas as pd
import matplotlib.pyplot as plt

from google.colab import files
uploaded = files.upload()


# Load agriculture crop production dataset
df = pd.read_csv('crop_yield.csv')

# Basic Statistics
print("\nBasic Statistics:")
print(df.describe())

# 1. Calculate total production of crops
total_production = df['Production'].sum()
print("\nTotal Production of Crops (in metric tons):", total_production)

# 2. Average yield of crops
average_yield = df['Yield'].mean()
print("Average Crop Yield (in metric tons per hectare):", average_yield)

# 3. Maximum area under cultivation
max_area = df['Area'].max()
print("Maximum Area Under Cultivation (in hectares):", max_area)

# 4. Minimum annual rainfall in the dataset
min_rainfall = df['Annual_Rainfall'].min()
print("Minimum Annual Rainfall (in mm):", min_rainfall)

# 5. Count of unique crops cultivated
unique_crops = df['Crop'].nunique()
print("Number of Unique Crops Cultivated:", unique_crops)



# 1. Plotting Crop Production by Year
plt.figure(figsize=(12, 6))
df.groupby('Crop_Year')['Production'].sum().plot(kind='bar', color='skyblue')
plt.title('Total Crop Production by Year')
plt.xlabel('Year')
plt.ylabel('Total Production (in metric tons)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

# 2. Crop Yield Distribution
plt.figure(figsize=(12, 6))
df['Yield'].plot(kind='hist', bins=30, color='lightgreen', edgecolor='black')
plt.title('Distribution of Crop Yield')
plt.xlabel('Yield (metric tons per hectare)')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()

# 3. Average Production by State
plt.figure(figsize=(12, 6))
df.groupby('State')['Production'].mean().sort_values().plot(kind='barh', color='orange')
plt.title('Average Crop Production by State')
plt.xlabel('Average Production (in metric tons)')
plt.ylabel('State')
plt.grid(axis='x')
plt.show()
```
