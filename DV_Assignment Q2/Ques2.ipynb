```python
import pandas as pd
import numpy as np

from google.colab import files
uploaded = files.upload()

# Load the TikTok dataset (replace the path with your file path)
df = pd.read_csv('tiktok_performance.csv')

# 1. Basic Operations

# a. Calculate the total number of likes for all videos
total_likes = df['Likes'].sum()
print("\nTotal Likes: ", total_likes)

# b. Find the average number of views
average_views = df['Views'].mean()
print("Average Views: ", average_views)

# c. Calculate the maximum and minimum video length
max_length = df['Video_Length'].max()
min_length = df['Video_Length'].min()
print(f"Max Video Length: {max_length}, Min Video Length: {min_length}")

# d. Filter videos that belong to the 'Comedy' category
comedy_videos = df[df['Category'] == 'Comedy']
print("\nComedy Videos:\n", comedy_videos)

# 2. Advanced Operations

# a. Sort videos by the number of likes in descending order
sorted_by_likes = df.sort_values(by='Likes', ascending=False)
print("\nVideos Sorted by Likes:\n", sorted_by_likes)

# b. Calculate the correlation between views and likes
correlation = df['Views'].corr(df['Likes'])
print("\nCorrelation between Views and Likes: ", correlation)

# c. Group by Category and calculate average likes and views
grouped = df.groupby('Category')[['Likes', 'Views']].mean()
print("\nAverage Likes and Views by Category:\n", grouped)

# d. Numpy operation: Calculate total sum of user followers
import numpy as np
total_followers = np.sum(df['User_Followers'])
print("\nTotal User Followers: ", total_followers)


```
